```python
# Task 2: Mark the success/failed launches for each site on the map
# Recreate the map centered at NASA JSC
nasa_coordinate = [29.559684888503615, -95.0830971930759]
site_map = folium.Map(location=nasa_coordinate, zoom_start=5)

# Optional: Add NASA JSC marker
site_map.add_child(folium.Circle(
    nasa_coordinate,
    radius=1000,
    color='blue',
    fill=True
).add_child(folium.Popup('NASA Johnson Space Center')))

site_map.add_child(folium.map.Marker(
    nasa_coordinate,
    icon=DivIcon(
        icon_size=(20,20),
        icon_anchor=(0,0),
        html='<div style="font-size: 12; color:blue;"><b>NASA JSC</b></div>'
    )
))

# Optional: Group launch markers into a cluster
marker_cluster = MarkerCluster().add_to(site_map)

# Add circles for each launch (green=success, red=failure)
for index, row in spacex_df.iterrows():
    coord = [row['Lat'], row['Long']]
    site_name = row['Launch Site']
    result = row['class']
    color = 'green' if result == 1 else 'red'
    label = 'Success' if result == 1 else 'Failure'

    folium.Circle(
        location=coord,
        radius=100,
        color=color,
        fill=True,
        fill_opacity=0.7,
        popup=f'{site_name} - {label}'
    ).add_to(marker_cluster)

site_map
```

```python
# TASK 3: Calculate the distances between a launch site to its proximities
import math

def haversine(lon1, lat1, lon2, lat2):
    R = 6371  # Earth radius in kilometers

    # Convert decimal degrees to radians
    lon1, lat1, lon2, lat2 = map(math.radians, [lon1, lat1, lon2, lat2])

    # Haversine formula
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = math.sin(dlat / 2)**2 + math.cos(lat1) * math.cos(lat2) * math.sin(dlon / 2)**2
    c = 2 * math.asin(math.sqrt(a))
    return R * c
nasa_lat, nasa_lon = 29.559684888503615, -95.0830971930759

# Add a new column with distance from NASA JSC
launch_sites_df['distance_from_nasa_jsc_km'] = launch_sites_df.apply(
    lambda row: haversine(nasa_lon, nasa_lat, row['Long'], row['Lat']),
    axis=1
)

launch_sites_df
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
from folium import Map, Marker, PolyLine
from folium.features import DivIcon

# Coordinates
launch_site = [28.5623, -80.5774]  # Cape Canaveral SLC-40

# Features around the site
features = {
    "Coastline": [28.56367, -80.57163],
    "Railway": [28.6100, -80.8070],
    "Highway": [28.5200, -80.7650],
    "City": [28.3200, -80.6076]
}

# Create map
site_map = Map(location=launch_site, zoom_start=10)

# Launch site marker
Marker(
    location=launch_site,
    popup="Launch Site: SLC-40",
    icon=folium.Icon(color='red', icon='rocket')
).add_to(site_map)

# Add feature markers and distances
for label, coord in features.items():
    # Calculate distance
    distance = calculate_distance(launch_site[0], launch_site[1], coord[0], coord[1])
    
    # Add feature marker
    Marker(
        location=coord,
        popup=label,
        icon=folium.Icon(color='blue', icon='info-sign')
    ).add_to(site_map)
    
    # Add distance label
    Marker(
        location=coord,
        icon=DivIcon(
            icon_size=(100, 20),
            icon_anchor=(0, 0),
            html=f'<div style="font-size: 12px; color:#d35400;"><b>{label}: {distance:.2f} KM</b></div>',
        )
    ).add_to(site_map)
    
    # Add line from launch site
    PolyLine([launch_site, coord], color='gray', weight=2).add_to(site_map)

site_map
```
